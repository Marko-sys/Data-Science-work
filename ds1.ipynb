```python
import pandas as pd
import requests
import json
import numpy as np

csv_path='convertcsv.csv'
df=pd.read_csv(csv_path, nrows=15)
df.head()
df_new = df.iloc[:, [0,1,2,3,4]]
df_new
print(df_new)
print("-----------------------------------------------------------------")
num_males =  len(df_new[df_new['gender'] == 'Male'])
num_females = len(df_new[df_new['gender'] == 'Female'])

print("M :",num_males)
print("F :",num_females)

import matplotlib.pyplot as plt

y=np.array([num_males,num_females])
mylabels=["Males", "Females"]
mycolours=["green","hotpink"]

plt.pie(y, labels= mylabels, colors=mycolours, autopct='%.1f%%', explode=(0,0.1), 
wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'}, textprops={'size': 'x-large', 'color':'white'}, shadow=True)

plt.title("Demographic")
plt.legend(title="GENDER:")
plt.show()
```
