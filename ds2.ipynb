```python
import pandas as pd
import requests
import urllib,json
from urllib.request import urlopen

csv_path='TECHM.NS.csv'
df=pd.read_csv(csv_path, nrows=10)
df.head
df_new = df.iloc[:, [1,2,3]]
df_new
print(df_new)
print("-----------------------------------------------------------------")

import matplotlib.pyplot as plt

h= df_new['High']
l= df_new['Low']

print("High :",h)
print("Low :",l)

diff=h-l
print(diff)

plt.plot(diff, linestyle='dotted')
plt.xlabel("No.of Days")
plt.ylabel("Difference Between High and Low")
plt.title("Tech Mahindra Stock - 10 Days ")
plt.show()
```
